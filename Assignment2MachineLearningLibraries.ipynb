### NumPy (Array Operations)

1.Creating Numpy array of shape(5,5) integer btw 1& 50

```python
import numpy as np

#creating Numpy array with randominteger from 1 to50
Array = np.random.randint(1,51,(5,5))
print(Array)

# adding each row (axis=1) adding across the row
Sum=np.sum(Array,axis=1)
print("Sum of rows",Sum)

#find which row has the maximum sum
Maximum_sum=np.argmax(Sum)
print(f"Row with maximum sum: {Array[Maximum_sum]} Index of this row is : {Maximum_sum}")

#find the mean value of the array
Array_mean=np.mean(Array)
print("Mean value of the array",Array_mean)

#replace the value greater than the mean with 0
Array[Array>Array_mean]=0
print("Updated array",Array)
```

###2. crate a NumPy array number 1 to 20 and take every 3rd element(tep=3) starting from 2nd element(index=1)


```python
import numpy as np

#creating array from 1 to 20
Array=np.arange(1,20)
print("Array:", Array)

# Slicing the array
Sliced_Array=Array[1::3]
print("Sliced Array:", Sliced_Array)
```

####Pandas (Data Handling)


 3.Load the dataset Iris data

```python
import pandas as pd
import seaborn as sns

#Loading the data set iris
iris=sns.load_dataset('iris')

#display 10 rows
print(iris.head(10))

#basic statistics
print(iris.describe())

#calculate sepal_length  median
median_sepal_length=iris['sepal_length'].median()
print("\nMedian of sepal_length:", median_sepal_length)

# Filter the DataFrame based on the conditions
Filter_iris= iris[(iris['sepal_length'] > median_sepal_length) & (iris['species'] == 'versicolor')]

# Print the filtered rows
print(Filter_iris)

#Grouping by species and computing the petal_length mean
Grouping=iris.groupby("species")["petal_length"].mean()
print(Grouping)
```

### 4.Add new column

```python
# adding new column petal_ratio
iris['petal_ratio'] =iris['petal_length'] / iris['petal_width']
print(iris.head())

# top 5rows with highest petal_ratio
petal_ratiohigh = iris.nlargest(5,'petal_ratio')
print(petal_ratiohigh)
```

### 5 Matplotlib  Histrogram and Scatter plot


### Histogram

```python
import matplotlib.pyplot as plt


# creating the histogram with 15 bin
plt.hist(x=iris["sepal_length"], bins=15, edgecolor="black")
plt.xlabel("Sepal Length")
plt.ylabel("Count")
plt.title("Histogram of Sepal Length")
plt.show()
```

###Scatter plot

```python
import matplotlib.pyplot as plt

# Create a scatter plot and color-coded by species.
plt.scatter(x=iris["sepal_length"], y=iris["petal_length"], c=iris["species"].astype("category").cat.codes,cmap="Wistia")
plt.colorbar(label="Species")


# Add labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot of Sepal Length vs Petal Length by Species")

#
plt.show()
```
